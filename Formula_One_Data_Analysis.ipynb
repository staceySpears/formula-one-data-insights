# Formula One Data Analysis

## Introduction
This notebook presents the analysis of Formula One data. We will load the cleaned data from Google Drive, perform exploratory data analysis (EDA), and visualize some key insights. The data includes information about races, drivers, constructors, results, standings, qualifying times, lap times, pit stops, seasons, circuits, status, sprint results, and constructor results.

```python
import os
import pandas as pd
from google.colab import drive

# Mount Google Drive to access files stored in Google Drive
drive.mount('/content/drive')

# Define the path to the folder containing the cleaned data
cleaned_data_dir = '/content/drive/My Drive/F1_cleaned_data'

# Verify the contents of the directory to ensure the files are present
print("Contents of the cleaned data directory:")
print(os.listdir(cleaned_data_dir))

# List of all cleaned CSV files with the correct filenames
csv_files = [
    'cleaned_races.csv', 'cleaned_drivers.csv', 'cleaned_constructors.csv', 'cleaned_results.csv',
    'cleaned_driver_standings.csv', 'cleaned_constructor_standings.csv', 'cleaned_qualifying.csv',
    'cleaned_lap_times.csv', 'cleaned_pit_stops.csv', 'cleaned_seasons.csv', 'cleaned_circuits.csv',
    'cleaned_status.csv', 'cleaned_sprint_results.csv', 'cleaned_constructor_results.csv'
]

# Dictionary to store DataFrames
dfs = {}

# Load data for all cleaned CSV files
for csv_file in csv_files:
    csv_path = os.path.join(cleaned_data_dir, csv_file)
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        # Store the DataFrame in the dictionary
        dfs[csv_file] = df
        print(f"Loaded {csv_file}")
    else:
        print(f"File not found: {csv_path}")
```
